## Importing Libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import export_text
import matplotlib.pyplot as plt
from sklearn import tree
import pandas as pd
import numpy as np

pd.set_option("max_columns", None)
```

## Reading the Data

```python
df_original = pd.read_csv("./archive/student-mat.csv")
df_original.head()
```

## Inspecting the Columns

```python
# list all columns available
print(list(df_original.columns))
```

```python
# print the count of values for each column
for col in df_original.columns:
    print("Column: ", col)
    print(df_original[col].value_counts(dropna=False).to_dict())
```

# 

# *Predicting: Internet Usage(without replacing no/yes with 0/1)*

```python
# copying df for the use
df = df_original.copy()
```

```python
# define the classification target
target = 'internet'
```

```python
# printing dataframe without print() function happen to be more beautiful and easy to read
df.head()
```

## Selecting X, y Datasets

```python
# selecting y dataset
y = df[target]
```

```python
# keep only numeric columns by checking the data type of each series
numeric_cols = [col for col in df.columns if type(df.loc[0, col]) is not str]        

# selecting X dataset
X = df[numeric_cols]
```

```python
X.head()
```

## Spliting the Datasets into Train/Test

```python
# split the dataset set into 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Building and Training the model

```python
# building the decision tree classifier model
classifier_model = tree.DecisionTreeClassifier()
```

## Training the model

```python
classifier_model = classifier_model.fit(X_train, y_train)
```

## Visualizing the Model

```python
# viewing the structure of the models as a decision tree
view_tree = tree.export_text(classifier_model, feature_names=list(X.columns))
print(view_tree)
```

```python
# improving the visualization with matplotlib and plot_tree function from scikitlearn
fig = plt.figure(figsize=(40,40))
_   = tree.plot_tree(classifier_model, feature_names=X.columns, filled=True)
```

## Predicting the Result

```python
preds = classifier_model.predict(X_test)
print(preds)
```

## Forming Accuracy Test on the Prediction

```python
print(accuracy_score(y_test, preds))
```

## Calculating Feature Importances

The code below uses the Gini importance score to calculate the importance of each feature for the final result.

Gini importance is computed as the sum of the number of splits where the feature was used divided by the total number of samples resulting in each split. The higher the score, the more important the feature.

```python
# sorting the feature importances in ascending order and printing in column-score relationship
for col, score in sorted(zip(X.columns, classifier_model.feature_importances_), key=lambda x: x[1]):
    print(f'Column: {col}\t\tScore: {score}')
```

## NOTE:

In the code above, we've performed the splitting and modelling processes in order to get the same reproducible result for each run, ``random_state`` parameter has been changed from ``None`` to `42` in each call.

## Testing Parameter Sensitivities

#### 1. splitter

**Hypothesis:** When its chosen random, we expect the accuracy to be lower than the default value, which is best.

```python
for i in range(10):
    print(f'Random state: {i}')
    preds_best = tree.DecisionTreeClassifier(random_state=i, splitter = 'best').fit(X_2_train, y_2_train).predict(X_2_test)
    print('Best split accuracy: ', accuracy_score(y_2_test, preds_best))
    preds_rndm = tree.DecisionTreeClassifier(random_state=i, splitter = 'random').fit(X_2_train, y_2_train).predict(X_2_test)
    print('Random split accuracy: ', accuracy_score(y_2_test, preds_rndm))
    print('\n')
```

Depending on the chosen ``random_state``, the results vary. In some cases, the `best` split achieves a better result than `random`, while in some cases it's otherwise.

#### 2. max_depth

**Hypothesis:** Since the tree is expanded to all nodes until pure leaves by default, the model can be learning noise and hence, can overfit. By limiting this parameter to a lower number, we can decrease the tendency to overfit and improve the accuracy score. 

```python
scores = []

for depth in range(2, 100):
    preds_max_depth = tree.DecisionTreeClassifier(random_state=42, max_depth = depth).fit(X_2_train, y_2_train).predict(X_2_test)
    scores.append(accuracy_score(y_2_test, preds_max_depth))
    
print('Best accuracy: ', np.max(scores))
plt.plot(range(2, 100), scores)
plt.xlabel('Value of max_depth')
plt.ylabel('Accuracy')
```

The model accuracy score stabilizes at around `max_depth = 15`. The resulting accuracy score is approximately the same as the original one (~53%).

#### 3. min_samples_split

**Hypothesis:** By increasing the min_samples_split value, we believe the tree might become too shallow and therefore, the accuracy might decrease. 

```python
scores = []

for samples in range(2, 100):
    preds_min_samples = tree.DecisionTreeClassifier(random_state=42, min_samples_split = samples).fit(X_2_train, y_2_train).predict(X_2_test)
    scores.append(accuracy_score(y_2_test, preds_min_samples))
    
plt.plot(range(2, 100), scores)
plt.xlabel('Value of min_samples')
plt.ylabel('Accuracy')
```

The model score varies a lot within different chosen ranges. This means the model acccuracy might be different if the range is higher. For example, the model accuracy score within `min_samples_split ~ 70-100` is appx. 60%.

#### 4. max_features

**Hypothesis:** By choosing `sqrt` as a max_features value, we believe the best split will not be found and might cause decrease in accuracy.

```python
preds_max_ftrs = tree.DecisionTreeClassifier(random_state=42, max_features='sqrt').fit(X_2_train, y_2_train).predict(X_2_test)
print(accuracy_score(y_2_test, preds_max_ftrs))
```

The resulting accuracy score is not much different from the one originally achieved.

#### 5. criterion

**Hypothesis:** Since the calculation methodologies are almost the same (`gini` and `entropy`), we believe choosing entropy might have negligible impact on the accuracy.

```python
preds_criterion = tree.DecisionTreeClassifier(random_state=42, criterion ='entropy').fit(X_2_train, y_2_train).predict(X_2_test)
print(accuracy_score(y_2_test, preds_criterion))
```

The resulting accuracy score is slightly higher than the one originally achieved. However, it has to be noted that the `entropy` method requires more computational power.

## Suggestions to Evaluate Model Performance

The model performance might be evaluated by the followings: 

 1. After building the model, we can find the accuracies on `X_test` and `X_valid` separately, and if the accuracy score achieved on the validation data is drastically higher than the accuracy score obtained on the test dataset, then its an indication of either overfitting or underfitting.
 2. Choosing multiple training/test/validation splits and comparing how the model performs with regards to different percentages can convey how stable the model predictions are. 
 3. We can evaluate the sub optimality of chosen parameters by testing the model with different parameter values and observing how the accuracy scores perform within each value range. For example, the accuracy score might decrease if a parameter value is chosen to be too large. Therefore, its important to test the optimality of parameters. 
 4. There might be features that might have negligible contribution to the model. In this case, to avoid noise coming from this feature (for example, the feature might be correlated with other features but in fact, carry no meaning for the final result), it should be dropped. 
 5. If the number of observations in the data is too little, underfitting or overfitting is inevitable.

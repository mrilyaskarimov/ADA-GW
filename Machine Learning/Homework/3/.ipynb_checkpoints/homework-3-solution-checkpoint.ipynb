## Encoding, Missing Data Imputation & Feature Reduction

We had two actions regarding missing values:

1) first we reduced # features where proportion of missing data points was more than 50% of total observations,
2) we dropped rows with more than 10 missing data points across the remaining features,
3) we imputed median and mode values for the rest of the missing data points based on the assumption that these will not skew the results much due to a lower number of missing data points.

### Functions

```python
def seperate_range(rng: str, lowest=False):
    if lowest:
        return min(map(int, rng.split("-")))
    return max(map(int, rng.split("-")))

def convert(ans: str):
    """
    Convert the string answers to ordinal numeric data type.
    """
    res = ans
    if ans in ['Yes', 'Always', 1]:
        res = 1
    elif ans in ["I don't know", "I'm not sure", "I am not sure", "Maybe", 'Sometimes', 0.5]:
        res = 0.5
    elif ans in ['No', 'Not eligible for coverage / N/A', 'Never', 0]:
        res = 0
        
    return res

def convert_levels(ans: str):
    """
    Convert the string answers to ordinal numeric data type.
    """
    res = ans
    if ans in ["Very difficult", "No, I don't think they would", "Never", "Not applicable to me", "Not open at all", "Not applicable to me (I do not have a mental illness)", "No, it has not", "No, they do not", "No, I don't think it would", " No, none did", "No, none did", "None did", "No", "None of them", "No, at none of my previous employers", "No, I only became aware later", "Male", "male", "Male ", "Man", "man", "M", "m", "sex is male", "Male."]:
        res = 0.0
    elif ans in ["Somewhat difficult", "Rarely", "Somewhat not open"]:
        res = 0.25
    elif ans in ["Maybe", "Maybe/Not sure", "Neutral", "Neither easy nor difficult", 'N/A (not currently aware)', "I don't know", "Sometimes", "N/A (not currently aware", "Female", "female", "Female ", "female/woman", " Female", "Woman", "I identify as female.", "woman", "F", "f"]:
        res = 0.5
    elif ans in ["Somewhat easy", "Often", "Somewhat open", "Some did", "Some of them", "I was aware of some", "Some of my previous employers"]:
        res = 0.75
    elif ans in ['Very easy', "Yes, I think they would", "Yes, I experienced", "Yes, I observed", "Very open", "Yes, they do", "Yes, it has", "Yes, they all did", "Yes, I was aware of all of them", "Yes, always", "Yes, all of them", "Yes, at all of my previous employers", "Yes, I think it would"]:
        res = 1.0
        
    return res

def convert_gender(ans: str):
    """
    Convert the string answers to ordinal numeric data type.
    """
    res = ans
    if ans in ["Male", "male", "Male ", "Man", "man", "M", "m", "sex is male", "Male."]:
        res = 0.0
    elif ans in ["Female", "female", "Female ", "female/woman", " Female", "Woman", "I identify as female.", "woman", "F", "f"]:
        res = 0.5
    else:
        res = 1.0
    return res
```

## Which Features to One-Hot Encode?

`country_of_work` or `country_of_residence` categories can be one hot encoded by having 2 major observations, such as the United States, The United Kingdom, and other countries as in `others` category, or by continent, such as Asia, Europe, etc.

`US_states` category can be also one hot encoded, by focusing on the major states and keeping other states in the `others` category.

`work_position` category can be one hot encoded by choosing 2-3 frequent work positions and  other positions as in âothersâ category. Meanwhile, we believe, we can one hot encode the work_position category by seniority of the work, as well - e.g. developer, supervisor, executive, etc. 
 
Given the lack of time, we did not spend efforts on finding key words in `physical_reason` and `mental_reason` categories. We might have done one hot encoded for those features as well, which could have yielded more insights.

## Which Features to Engineer?

The following feature combinations might be useful in modelling:

- have mental health issue / size company (how company size affects whether the person experiences mental health issues)
- working remotely * mental health issue (how mental health issues affect the decision to work remotely)
- has family history * has mental issue (how genes affect mental health)
- would hurt career if revealed mental health issue * self employed (how interacted the two are)
- gender * would share mental health issue (how gender affects the decision to share and discuss issues)
- previous experiences * would discuss mental heatlh issues now (how the past affects the present)
- age * has formally discussed mental health issue (how the two interact togetehr)
- mental health benefits provided * has mental health issue (weighted mental health benefits by those who actually suffer from it)
- would observe negative conseq in coworkers * willingness to discuss with employers and other coworkers, supervisor (how observations affect willingness to discuss)
